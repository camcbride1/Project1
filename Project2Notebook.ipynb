# Power Company Data Recall and Bill View Application

## Functions:

```python
import random

# Generates user data for project use.
def genUserData(): 
    ids = list([random.randrange(100,999) for index in range(10)])
    powerHour = list([random.randrange(300,900) for index in range(10)])
    returnDict = dict()
    for index in range(10):
        returnDict[ids[index]] = powerHour[index]
    return returnDict

# Displays user data 
def displayUserData(dictionary, powerCost):
    strList = []
    outStr = ''
    for item in dictionary:
        strList.append(f"\nID number: {item}, Power Usage in kWh: {dictionary[item]}, Bill: {round(dictionary[item] * powerCost, 2)}")
    for item in strList:
        outStr += (item + '\n')
    return outStr

# Returns the sum off all bills
def viewTotalPower(data, cost):
    returnVal = 0
    lyst = data.values()
    cost = float(cost)
    for item in lyst:
        returnVal += float(item)

    return returnVal * cost

# Changes the cost of power
def changePowerCost():
    while True:
        print("\nEnter a cost for power: ")
        userInput = input()
        try:
            userInput = float(userInput)
        except:
            print("\nNot a valid input")
        else:
            break
    return float(format(userInput, '.2f'))

# Main application       
def mainApp():
    userData = genUserData()
    powerCost = .10
    while True:
        print("\nChoose an Option:\n1) Customer power usage and bill\n2) View total power usage and cost\n3) Change power cost per kWh\n9) Exit Program")
        
        userInput = input()
        if userInput == '1':
            print(displayUserData(userData, powerCost))
        elif userInput == '2':
            print(f"\nTotal power bill is {round(viewTotalPower(userData, powerCost), 2)}\nPrice per kWh is {powerCost}\n")
        elif userInput == '3':
            powerCost = changePowerCost()
        elif userInput == '9':
            break
        else:
            print("\nNot an accepted value")

# Driver
mainApp()
```
